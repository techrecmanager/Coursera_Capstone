### Part 1: Scraping Canadian postal codes (that begin with M), located within the city of Toronto in the province of Ontario from wikipedia

<strong><em>Explore and cluster the neighborhoods in Toronto.</em></strong>

Scraping list of postal codes in Canada where the first letter is M and are located within the city of Toronto in the province of Ontario. <br>

The following code will scrape Wikipedia page "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

Data from wikipedia canadian postal codes table will be extracted and transformed into pandas dataframe.

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

### Convert HTML table to Pandas dataframe

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

### Data Cleanup and re-grouping.
The scraped wikipedia table contains some un-wanted entries and needs some cleanup.
The following tasks will be performed:
* Drop/ignore cells with un-assigned boroughs.
* Combine neighbourhoods with same Postalcode, separated with a 'comma'.
* If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
# Drop/ignore cells with un-assigned boroughs

df1 = df[df.Borough != 'Not assigned']
```

```python
df1.head()
```

```python
# Rename column name 'Postal Code' to 'Postcode'

df1.rename(columns={'Postal Code':'Postcode'},inplace=True)

df1.head()
```

```python
# Combine neighbourhoods with same Postalcode, separated with a 'comma'.

df2 = df1.groupby(['Postcode','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)
```

```python
df2.head()
```

```python
# If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough. 
#This means replacing the name of the neighbourhoods which are 'Not assigned' with names of their Borough

df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2
```

```python
# use the .shape method to print the number of rows in the dataframe.

df2.shape
```

### Part 2: Fetch the latitude and the longitude coordinates of each neighborhood.

```python
import random 
from geopy.geocoders import Nominatim 
from IPython.display import Image 
from IPython.core.display import HTML 
from IPython.display import display_html
from pandas.io.json import json_normalize
```

```python
# Importing the csv file that contains latitudes and longitudes for neighbourhoods in Toronto

coordinates = pd.read_csv('https://cocl.us/Geospatial_data')
```

```python
coordinates.head()
```

```python
# Rename column name 'Postal Code' to 'Postcode'

coordinates.rename(columns={'Postal Code':'Postcode'},inplace=True)
```

```python
coordinates.head()
```

```python
# Merge the two DFs (on "Postcode" column) to map Latitudes and Longitudes for neighbourhoods in Toronto

df3 = pd.merge(df2,coordinates,on='Postcode')
df3.head()
```

### Part 3: Explore and cluster boroughs that contain the word Toronto

```python
# Fetch all rows from the data frame which have Toronto as the "Borough" value.

df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
from sklearn.cluster import KMeans
```

```python
# let's visualize toronto neighborhoods and how they cluster together using FOLIUM

map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
# Lets figure out the clustering of Toronto neighborhoods using KMeans clustering 

k=5
toronto_clustering = df4.drop(['Postcode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df4.head()
```

```python
# create a map using FOLIUM

map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)
```

```python
# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
